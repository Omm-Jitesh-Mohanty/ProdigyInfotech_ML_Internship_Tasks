```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
train = pd.read_csv('C:/Users/ommji_mttma5p/OneDrive/Desktop/Prodigymltask1/train.csv')
train.head()
```

```python
train.info()
```

```python
train.isnull().sum()
```

```python
test = pd.read_csv('C:/Users/ommji_mttma5p/OneDrive/Desktop/Prodigymltask1/test.csv')
test.head()
```

```python
sample = pd.read_csv('C:/Users/ommji_mttma5p/OneDrive/Desktop/Prodigymltask1/sample_submission.csv')
sample.head()
```

```python
X_train = train[['GrLivArea', 'BedroomAbvGr', 'FullBath']]
y_train = train['SalePrice']

X_test = test[['GrLivArea', 'BedroomAbvGr', 'FullBath']]
y_test = sample['SalePrice']
```

```python
corr_matrix = X_train.corr()

plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of Train Data')
plt.show()
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error (MSE):", mse)
```

```python
mae = mean_absolute_error(y_test, y_pred)
print("Mean Absolute Error (MAE):", mae)
```

```python
r2 = r2_score(y_test, y_pred)
print("R-squared (R2) Score:", r2)
```

```python
rmse = np.sqrt(mse)
print("Root Mean Squared Error (RMSE):", rmse)
```

```python
plt.figure(figsize=(35, 10))
plt.plot(y_test, label='Actual SalePrice', color='blue')
plt.plot(y_pred, label='Predicted SalePrice', color='red')

plt.xlabel('Data Points')
plt.ylabel('SalePrice')
plt.title('Actual vs. Predicted SalePrice')
plt.legend()
plt.show()
```
